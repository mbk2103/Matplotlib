```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numArray = np.linspace(0, 15, 25)
```

```python
numArray
```

```python
numArray2 = numArray ** 2
```

```python
(myFigure, myAxes) = plt.subplots()
myAxes.plot(numArray, numArray2, "r")
```

```python
myAxes 
```

```python
(myFigure, myAxes) = plt.subplots()
myAxes.plot(numArray, numArray2, "r")
myAxes.plot(numArray2 , numArray, "#0009FF") # Look at rgb hex code
```

```python
(myFigure, myAxes) = plt.subplots()
myAxes.plot(numArray, numArray2, "r", alpha = 1)
myAxes.plot(numArray2 , numArray, "#0009FF", alpha = 0.5) 
```

```python
(newFigure,newAxes) = plt.subplots()
newAxes.plot(numArray, numArray + 2, color = "green", alpha = 1)
newAxes.plot(numArray, numArray + 4, color = "gray", alpha = 0.8)
```

```python
(newFigure,newAxes) = plt.subplots()
newAxes.plot(numArray, numArray + 1, color = "green", linewidth = 1)
newAxes.plot(numArray, numArray + 9, color = "gray", linewidth = 5)

newAxes.plot(numArray, numArray + 3, color = "pink", linewidth = 2, linestyle = "-.")
newAxes.plot(numArray, numArray + 5, color = "blue", linewidth = 3, linestyle = ":")
newAxes.plot(numArray, numArray + 7, color = "purple", linewidth = 4, linestyle = "--")

newAxes.plot(numArray, numArray + 12, color = "red", linewidth = 1, linestyle = "--", marker = "o", markerfacecolor = "white", markersize = 8)
newAxes.plot(numArray, numArray + 15, color = "yellow", linewidth = 1, linestyle = "--", marker = "*", markerfacecolor = "black", markersize = 15)
```
